```python
import matplotlib.pyplot as plt
# 👩‍🎓 Student Data
students = ["Ali", "Sara", "Ahmed", "Ayesha", "Bilal"]

math = [78, 85, 60, 90, 70]
science = [82, 88, 65, 95, 75]
english = [75, 80, 70, 85, 68]


# 📈 1. Simple Line Plot

plt.plot(students, math, label="Math")
plt.plot(students, science, label="Science")
plt.plot(students, english, label="English")

plt.title("Student Marks (Line Chart)")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.legend()
plt.show()


# 📊 2. Simple Bar Chart
x = range(len(students))

plt.bar(x, math, label="Math")
plt.bar(x, science, bottom=math, label="Science")

plt.xticks(x, students)

plt.title("Marks Comparison (Bar Chart)")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.legend()
plt.show()



# 📉 3. Simple Histogram
all_marks = math + science + english

plt.hist(all_marks, bins=5)

plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Count")

plt.show()


# 📍 4. Simple Scatter Plot
plt.scatter(math, science)

plt.title("Math vs Science")
plt.xlabel("Math Marks")
plt.ylabel("Science Marks")

plt.show()
```
